# Importing required libraries

```python
import json
from tkinter import *
from tkinter import messagebox
```

# Function for validating login credentials 

```python
def login():
    #getting form data
    uname=username.get()
    pwd=password.get()
    #applying validation
    if uname=='' or pwd=='':
        message.set("fill the empty field!!!")
    else:
        if uname=="admin" and pwd=="admin":
            login_screen.destroy()
            root.deiconify()
        else:
            message.set("Wrong username or password!!!")
```

# Creating Login window

```python
root=Tk()
root.title("Inventory Management System")
root.geometry("1000x500")
root.withdraw()
root.config(bg="light blue")


login_screen = Toplevel()
    
login_screen.title("Inventory management system account Login")
    
login_screen.geometry("300x250")
    
username = StringVar()
password = StringVar()
message=StringVar()
    
Label(login_screen,width="300", text="Please enter details below", bg="orange",fg="white").pack()
   
Label(login_screen, text="Username * ").place(x=20,y=40)
    
Entry(login_screen, textvariable=username).place(x=90,y=42)
    
Label(login_screen, text="Password * ").place(x=20,y=80)

Entry(login_screen, textvariable=password ,show="*").place(x=90,y=82)

Label(login_screen, text="",textvariable=message).place(x=95,y=100)

Button(login_screen, text="Login", width=10, height=1, bg="orange",command=login).place(x=105,y=130)
#login_screen.mainloop()

```

```python
top=Frame(root,bg='sky blue', height=300, width=1000)
top.grid(row=0, column=0)

bottom = Frame(root, bg='blue', height=200, width=1000)
bottom.grid(row=1, column=0)
```

## When refresh button is pressed, it will show the updated inventory after adding/updating/deleting products

```python
import tkinter.scrolledtext as st
fd = open("record.json",'r')
r = fd.read()
fd.close()
record = json.loads(r)
Heading=Label(top, text="Monica's Store ", font=('arial 40 bold'), bg='sky blue' )
Heading.place(x=300,y=0)
a1=Label(top, text="Product ID", bg='sky blue')
a1.place(x=0,y=80)
    
a2=Label(top, text="Product name", bg='sky blue')
a2.place(x=100,y=80)
    
a3=Label(top, text="Product category", bg='sky blue')
a3.place(x=300,y=80)
    
a4=Label(top, text="Brand", bg='sky blue')
a4.place(x=430,y=80)
    
a5=Label(top, text="Price in Rs", bg='sky blue')
a5.place(x=580,y=80)
    
a6=Label(top, text="Quantity", bg='sky blue')
a6.place(x=730,y=80)

data=st.ScrolledText(top, height=10, width=120)
data.insert(END,'\n'.join("{:10} {:25} {:15} {:15} {:10} {:15}".format(key,values['name'],values['category'], values['brand'], 

                                                                values['pr'], values['qn']) for key,values in record.items()))

def refresh():
    data.delete("1.0", "end")
    data.insert(END,'\n'.join("{:10} {:25} {:15} {:15} {:10} {:15}".format(key,values['name'],values['category'], values['brand'], 

                                                                values['pr'], values['qn']) for key,values in record.items()))

ref=Button(top, text="Refresh inventory",command=refresh, bg='blue')
ref.place(x=400,y=270)
data.place(x=0,y=100)

```

# Function for displaying sales transactions

```python
def sale():
    show=Toplevel()
    show.title("Sales transactions")
    show.geometry("1000x300")
    show.config(bg="light blue")
    fd = open("sales.json",'r')
    r = fd.read()
    fd.close()
    sales = json.loads(r)
    b1=Label(show, text="Transaction ID", bg='sky blue').place(x=0,y=80)
    b2=Label(show, text="Product ID", bg='sky blue').place(x=98,y=80)
    b3=Label(show, text="Product quantity", bg='sky blue').place(x=230,y=80)
    b4=Label(show, text="Amount in Rs", bg='sky blue').place(x=350,y=80)
    b5=Label(show, text="Billing Time", bg='sky blue').place(x=570,y=80)
    
    sale_data=st.ScrolledText(show, height=10, width=120)
    sale_data.insert(END, '\n'.join("   {:10} {:10} {:10} {:15}             {:20}".format(key,values['prod'],
                                                                                  values['qn'], values['amount'], values['Billing Time']) for key,values in sales.items()))
    sale_data.place(x=0,y=100)
    show.mainloop()


```

```python
def add():
    add.add_inv=Toplevel()
    add.add_inv.title("Add a product to inventory")
    add.add_inv.geometry("500x500")
    add.add_inv.config(bg="sky blue")

    add.id=Label(add.add_inv, text="Enter product Id:", bg='sky blue')
    add.id.place(x=10, y=50)

    add.get_id=Entry(add.add_inv)
    add.get_id.place(x=200, y=50)

    add.prod_name=Label(add.add_inv, text="Enter product name:", bg='sky blue')
    add.prod_name.place(x=10, y=80)

    add.get_name=Entry(add.add_inv)
    add.get_name.place(x=200, y=80)

    add.prod_cat=Label(add.add_inv, text="Enter product category:", bg='sky blue')
    add.prod_cat.place(x=10, y=110)

    add.get_cat=Entry(add.add_inv)
    add.get_cat.place(x=200, y=110)

    add.prod_brand=Label(add.add_inv, text="Enter product brand:", bg='sky blue')
    add.prod_brand.place(x=10, y=140)

    add.get_brand=Entry(add.add_inv)
    add.get_brand.place(x=200, y=140)

    add.prod_price=Label(add.add_inv, text="Enter price of product:", bg='sky blue')
    add.prod_price.place(x=10, y=170)

    add.get_price=Entry(add.add_inv)
    add.get_price.place(x=200, y=170)

    add.prod_qn=Label(add.add_inv, text="Enter product quantity:", bg='sky blue')
    add.prod_qn.place(x=10, y=200)

    add.get_qn=Entry(add.add_inv)
    add.get_qn.place(x=200, y=200)
    
    add.prod_add=Button(add.add_inv, text="Add product", command=add_prod, bg='blue')
    add.prod_add.place(x=100, y=230)
```

# Function for validating input values and then adding to inventory
## It will display an error message if already existing product ID is entered

```python
def add_prod():    
    prod_id = add.get_id.get()
    name = add.get_name.get()
    category=add.get_cat.get()
    brand=add.get_brand.get()
    pr = add.get_price.get()
    qn = add.get_qn.get()
    try:
        i1 = int(prod_id)
        i2 = int(pr)
        i3 = int(qn)
    except ValueError:
        messagebox.showerror("Error", "Only integers are allowed for Product ID, price and quantity")
    else:
        if prod_id=='':
            messagebox.showerror("Error", "Please enter Product ID")
        elif prod_id in record.keys():
            messagebox.showinfo("Information", "This product ID already exists. Please enter another product ID")
        elif name=='':
            messagebox.showerror("Error", "Please enter Product name")
        elif category=='':
            messagebox.showerror("Error", "Please enter Product category")
        elif brand=='':
            messagebox.showerror("Error", "Please enter Brand")
        elif pr=='':
            messagebox.showerror("Error", "Please enter Price")
        elif qn=='':
            messagebox.showerror("Error", "Please enter Product quantity")
        else:        
            record[prod_id] = {'name':name,'category':category,'brand':brand, 'pr': int(pr), 'qn': int(qn)  }

            js = json.dumps(record)

            fd = open("record.json",'w')
            fd.write(js)
            fd.close()
            add.add_inv.destroy()
            msg.set("Product added Successfully!!!")
```

# Function for validating input values and then deleting the product

```python
def delete_prod():
    delete_id=delete.del_prod_id.get()
    try:
        i1 = int(delete_id)
    except ValueError:
        messagebox.showerror("Error", "Only integers are allowed for Product ID")
    else:
        if delete_id=='':
            messagebox.showerror("Error", "Please enter Product ID")
        elif delete_id not in record.keys():
            messagebox.showinfo("Information", "No such product in inventory")
        else:
            del record[delete_id]

            js = json.dumps(record)
            fd = open("record.json",'w')
            fd.write(js)
            fd.close()
            delete.del_prod.destroy()
            msg.set("Product deleted successfully!!!")
```
